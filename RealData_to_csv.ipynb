```python
data = pd.read_excel('blouse1.xlsx')
data.head(5)
```

```python
reference_all = data[['title','contents','style_1','style_2','style_3','style_4']].values.tolist()
reference_s1 = data[['title','contents','style_1']].values.tolist()
reference_s2 = data[['title','contents','style_2']].values.tolist()
reference_s3 = data[['title','contents','style_3']].values.tolist()
reference_s4 = data[['title','contents','style_4']].values.tolist()


tc=data[['title','contents']]
tc=tc.values.tolist()
tc
```

```python
import sister
```

```python
sentence_embedding = sister.MeanEmbedding(lang="file",fasttextfile="cc.ko.10.bin")
```

```python
def merge_ti_con(result_list):
    vector1_list = []
    vector2_list = []
    
    for i in range(len(result_list)): 
    
        s1 = sentence_embedding(result_list[i][0])
        s2 = sentence_embedding(result_list[i][1])
        vector1_list.append(s1)
        vector2_list.append(s2)
        
        
    vector1_df= pd.DataFrame(vector1_list) #DataFrame으로
    vector2_df = pd.DataFrame(vector2_list) 
    
    return pd.concat([vector1_df,vector2_df],axis= 1) #merge
```

```python
result_df=merge_ti_con(tc)
```

```python
style_1=data[['style_1']]
style_2=data[['style_2']]
style_3=data[['style_3']]
style_4=data[['style_4']]

style_1
```

```python
def style1_conv(x):
    if x == '격식차린':
        return 0
    elif x == '바캉스룩':
        return 1
    elif x == '야외활동':
        return 2
    elif x == '이국적인':
        return 3
    elif x == '이지웨어':
        return 4
    else :
        return 5    

def style2_conv(x):
    if x == '스포티한':
        return 0
    elif x == '일상복':
        return 1
    else :
        return 2

def style3_conv(x):
    if x == '걸리쉬':
        return 0
    elif x == '매니쉬':
        return 1
    elif x == '섹시한':
        return 2
    elif x == '우아한':
        return 3
    else :
        return 4

def style4_conv(x):
    if x == '유니크한':
        return 0
    elif x == '장식이 없는':
        return 1
    else :
        return 2   
```

```python
style_1['style_1']= style_1['style_1'].apply(style1_conv)

style_2['style_2']= style_2['style_2'].apply(style2_conv)
style_3['style_3']= style_3['style_3'].apply(style3_conv)
style_4['style_4']= style_4['style_4'].apply(style4_conv)

style_1
```

```python
normalize_style1 = (style_1['style_1'] - style_1['style_1'].min()) / (style_1['style_1'].max() - style_1['style_1'].min()) # style_1 정규화
normalize_style2 = (style_2['style_2'] - style_2['style_2'].min()) / (style_2['style_2'].max() - style_2['style_2'].min()) # style_2 정규화
normalize_style3 = (style_3['style_3'] - style_3['style_3'].min()) / (style_3['style_3'].max() - style_3['style_3'].min()) # style_3 정규화
normalize_style4 = (style_4['style_4'] - style_4['style_4'].min()) / (style_4['style_4'].max() - style_4['style_4'].min()) # style_4 정규화

nor_style = [normalize_style1,normalize_style2,normalize_style3,normalize_style4]
```

```python
normalize_style = []

for i in range(len(nor_style)):
        normalize_style.append(pd.DataFrame(nor_style[i]))
        
final_result=pd.concat([result_df,normalize_style[0],normalize_style[1],normalize_style[2],normalize_style[3]],axis= 1)

final_result
```
